```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# 生成网格点
xx, yy = np.meshgrid(np.linspace(-3, 3, 500), np.linspace(-3, 3, 500))
np.random.seed(0)

# 生成异或数据
X = np.random.randn(300, 2)
Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0) # 逻辑异或操作

plt.scatter(X[:, 0], X[:, 1], )
```

```python
clf = svm.NuSVC(gamma='auto') # NuSVC可以调整支持向量个数
clf.fit(X, Y)
```

```python
# Evaluate the decision function for the samples in X.
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]) # c_ -> column
# np.c_[xx.ravel(), yy.ravel()]
print(Z.shape)
Z = Z.reshape(xx.shape)
```

```python
# 参考：https://www.jb51.net/article/193182.htm
plt.imshow(
    Z, 
    interpolation='nearest', # 热图的显示形式, nearest和none用的比较多
    extent=( xx.min(), xx.max(), yy.min(), yy.max() ), # 两个坐标轴的范围
    aspect='auto', # 指定热图的单元格大小，默认为equal，'auto'根据画布的大小动态调整单元格的大小
    origin='lower', # 原点位置，默认为upper，在坐标轴右上角，lower的origin在坐标轴左下角
    cmap=plt.cm.PuOr_r # colormap 指定渐变色，默认viridis（翠绿色）
    # alpha, 透明度
)
# 等高线
# levels 指定该第几条等高线
contours = plt.contour(xx, yy, Z, levels=[0], linewidths=2, linestyles='dashed')

# 描点
# c->class
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired, edgecolors='k')
plt.xticks()
plt.yticks()
plt.axis([-3, 3, -3, 3])
plt.show()
```
