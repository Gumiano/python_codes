```python
import numpy as np
import matplotlib.pyplot as plt

beg = -5
end = 5
x = np.linspace(beg, end, 1000)
def calc_basic_term(n):
    return x ** n

def plot_multi_curves(max_n=8, n_row=2, n_col=4, func=calc_basic_term):
    fig, subplots = plt.subplots(nrows=n_row, ncols=n_col, figsize=(16, 9))
    for r in range(n_row):
        for c in range(n_col):
            # 0, 0 -> x^0
            # 0, 1 -> x^1
            # 1, 0 -> x^5
            y_data = func(r*n_col+c)
            ax = subplots[r][c] # 获取子图

            ax.plot(x, y_data, color='black')
            
            # 配置
            max_y = max(y_data)
            ax.set_ylim(-max_y * 1.1, max_y * 1.1)
            ax.axvline(0, ymin=0, linestyle='dashed', color='black')
            ax.axhline(0, xmin=0, linestyle='dashed', color='black')

    plt.show()

plot_multi_curves()
```
