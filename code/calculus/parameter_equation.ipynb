```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 2, 1000)
x = 4 * (t ** 2) - 4
y = 2 * t - 2 * (t ** 3)

x1 = x.copy()
x1.sort()
x1 = x1[::100]
y1 = x1 / 2
y2 = -1 * y1

ax = plt.gca()
ax.plot(x, y)
ax.plot(x1, y1, 'g-.')
ax.plot(x1, y2, 'g-.')

plt.plot()
```
