```python
import pandas as pd

dish_profit = '../data/chapter3/demo/data/catering_dish_profit.xls'
data = pd.read_excel(dish_profit, index_col='菜品名')
# data = data['盈利'].copy()
data = data['盈利']
data.sort_values(ascending=False)
data.head()
```

```python
# 帕累托图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
data.plot(kind='bar', color='blue')
plt.ylabel('盈利（元）') # 第一个y label

p = 1.0 * data.cumsum() / data.sum() # 累计百分比
print(p)
p.plot(color='r', secondary_y=True, style='-o', linewidth=2) # 第二个图形，并设置第二个y轴，共享x轴
# 给85%的地方加上注释
plt.annotate( format(p[6], '.4%'), xy=(6, p[6]), xytext=(6*0.9, p[6]*0.9), arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2") )

plt.ylabel('盈利（比例）') # 第二个y label
plt.show()
```
