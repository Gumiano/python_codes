```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
# 正常用户
df_normal = pd.read_csv('../data/chapter3/demo/data/user.csv')
df_normal.head()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df_normal['Date'], df_normal['Eletricity'])
plt.xlabel('日期')
# x ticks
x_major_locator = plt.MultipleLocator(7) # 根据周期获取刻度
ax = plt.gca() # 获取当前坐标轴 get current axis
ax.xaxis.set_major_locator(x_major_locator)

plt.ylabel('每日电量')
plt.title('正常用户电量趋势')
plt.show()
```

```python
# 窃电用户
df_steal = pd.read_csv('../data/chapter3/demo/data/Steal user.csv')
df_steal.head()
```

```python
plt.figure(figsize=(10, 9))
plt.plot(df_steal['Date'], df_steal['Eletricity'])

plt.xlabel('日期')
plt.ylabel('每日电量')
# x ticks
x_major_locator = plt.MultipleLocator(7)
ax = plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.title('窃电用户每日电量趋势')
plt.show()
```
