```python
import pandas as pd
import numpy as np

np.random.seed(seed=0)
matrix = np.random.randn(6, 5)
df = pd.DataFrame(matrix)
sequence = pd.Series(range(1, 101))

print(df)
```

## 累计函数

```python
print('-------------------------------------------------------')
# np.cumsum(range(1, 101))
print(df.cumsum())
print(sequence.cumsum())

print('-------------------------------------------------------')
# np.cumprod(range(1, 6))
print(df.cumprod())
```

```python
print('-------------------------------------------------------')
print(df.cummax()) 

print('-------------------------------------------------------')
print(df.cummin()) 
```

## 滑动窗口

```python
print('-------------------------------------------------------')
# df.rolling(window=3).sum()

print('-------------------------------------------------------')
print(df.rolling(window=3).aggregate(lambda x : x.prod()))
print(df.rolling(window=3).apply(lambda x : x.prod()))
```

```python
df.rolling(window=3).std()
```
