## 曲线

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(seed=0)
x = np.arange(-50, 51)
y = x * x + np.random.normal()
plt.scatter(x, y)

ax = plt.gca()
# plt.axvline(x=0, ymin=0, ymax=2500) # 添加一条垂线
ax.vlines([-20, 20], ymin=0, ymax=2500, linestyles='dashed', colors='red') # 添加多条垂线
ax.set_ylim(0, 2500)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.plot(x, y, 'bp--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

x = pd.Series(np.exp(np.arange(20)))
plt.figure(figsize=(8, 9))
ax1 = plt.subplot(2, 1, 1)
x.plot(label='原始数据图', legend=True)

ax2 = plt.subplot(2, 1, 2)
x.plot(logy=True, label='原始数据图', legend=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

error = np.random.randn(10)
y = pd.Series(np.sin(np.arange(10)))
y.plot()
plt.show()
y.plot(yerr=error) # 误差棒图
plt.show()
```

## 饼图

```python
import matplotlib.pyplot as plt
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10] # 每一块的比例
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # 突出某些块

# autopct 显示比例
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

## 箱型图

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.random.randn(1000)
D = pd.DataFrame([x, x+1]).T
D.plot(kind='box')
plt.show()
```
