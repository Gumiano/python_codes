```python
import pandas as pd
import numpy as np

catering_sale = '../data/chapter3/demo/data/catering_fish_congee.xls'
data = pd.read_excel(catering_sale, names=['date', 'sale']) # names指定列名
```

```python
groups = 8
bins = [i * 500 for i in range(groups+1)]
labels = [f'[{i*500},{(i+1)*500})' for i in range(groups)]

data['sale分层'] = pd.cut(data.sale, bins, labels=labels)
aggResult = data.groupby(by=data['sale分层']).agg({'sale': np.size}) # 对sale分层统计
pAggResult = round(aggResult/aggResult.sum(), 2) * 100 # percentage
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
pAggResult['sale'].plot(kind='bar', width=0.8, fontsize=10, color='r')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('季度销售额频率分布直方图', fontsize=20)
plt.show()
```
